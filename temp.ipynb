```python
# %%
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = "vscode"

# 1. npy 파일 로드
pc = np.load("output.npy")

print("shape:", pc.shape)
print("dtype:", pc.dtype)

# 2. shape 정리
# 예상: [N, 3] 또는 [1, N, 3]
if pc.ndim == 3:
    pc = pc[0]

if pc.shape[-1] != 3:
    raise ValueError(f"Expected point cloud shape [N, 3], but got {pc.shape}")

# 3. Plotly 3D 시각화
fig = go.Figure(data=[
    go.Scatter3d(
        x=pc[:, 0],
        y=pc[:, 1],
        z=pc[:, 2],
        mode="markers",
        marker=dict(size=2),
        name="point cloud"
    )
])

fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z",
        aspectmode="data"
    ),
    width=800,
    height=700
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = "vscode"

pc = np.load("hand_aggregated.npy", allow_pickle=True)

if pc.ndim == 3:
    pc = pc[0]

if pc.shape[-1] != 3:
    raise ValueError(f"Expected point cloud shape [N, 3], but got {pc.shape}")

print("shape:", pc.shape)
print("min:", pc.min(axis=0))
print("max:", pc.max(axis=0))
print("range:", pc.max(axis=0) - pc.min(axis=0))

# 중심과 동일 축 범위 계산
center = pc.mean(axis=0)
max_range = (pc.max(axis=0) - pc.min(axis=0)).max() / 2

x_range = [center[0] - max_range, center[0] + max_range]
y_range = [center[1] - max_range, center[1] + max_range]
z_range = [center[2] - max_range, center[2] + max_range]

fig = go.Figure(data=[
    go.Scatter3d(
        x=pc[:, 0],
        y=pc[:, 1],
        z=pc[:, 2],
        mode="markers",
        marker=dict(size=2),
        name="hand point cloud"
    )
])

fig.update_layout(
    scene=dict(
        xaxis=dict(title="X", range=x_range),
        yaxis=dict(title="Y", range=y_range),
        zaxis=dict(title="Z", range=z_range),
        aspectmode="cube"
    ),
    width=800,
    height=700
)

fig.show()
```

```python
# %%
import numpy as np
import os

# 손 데이터 파일 구조 확인
data_dir = 'DexGraspNet/data/dexgraspnet'
npy_files = sorted([f for f in os.listdir(data_dir) if f.endswith('.npy')])[:3]

for fname in npy_files:
    fpath = os.path.join(data_dir, fname)
    data = np.load(fpath, allow_pickle=True)
    print(f"\n{fname}")
    print(f"  shape: {data.shape}")
    print(f"  dtype: {data.dtype}")
    print(f"  ndim: {data.ndim}")
    
    if data.ndim == 1 and len(data) > 0:
        print(f"  First element type: {type(data[0])}")
        elem = data[0]
        if isinstance(elem, np.ndarray):
            print(f"  First element shape: {elem.shape}")
            print(f"  First element dtype: {elem.dtype}")
            print(f"  First element sample: {elem.flat[:6]}")
        else:
            print(f"  First element: {elem}")
```
